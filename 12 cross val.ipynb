```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('modelling_mpg', data_home='Datasets')
df.head()
```

```python
X = df.drop(['mpg'], axis=1)
y = df['mpg']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.intercept_
```

```python
coef_parem = pd.DataFrame(model.coef_, index=X.columns, columns=['Coefficent'])
coef_parem
```

```python
y_pred = model.predict(X_test)
```

```python
dict = {'Actual : ': y_test, ' Predict : ': y_pred}
compare = pd.DataFrame(dict)
compare.sample(8)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
def evualvation_metrics(y_test, y_pred):
    SCORE = r2_score(y_test, y_pred)
    MSE = mean_squared_error(y_test, y_pred)
    MAE = mean_absolute_error(y_test, y_pred)
    RMSE = np.sqrt(mean_squared_error(y_test, y_pred))
    return print(f" MSE = {MSE}\n RMSE = {RMSE}\n MAE = {MAE}\n Score = {SCORE}")
```

```python
evualvation_metrics(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
acc = cross_val_score(estimator=model, X=X_train, y=y_train, cv=10)
acc
```
