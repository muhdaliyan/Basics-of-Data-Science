```python
df = pd.read_csv('Datasets/USArrests.csv', index_col=0)
df.head()
```

```python
from pyclustertend import hopkins
from sklearn.preprocessing import scale
hopkins(scale(df), df.shape[0])
```

K means Modelling 

```python
from sklearn.cluster import KMeans
```

```python
k_means = KMeans(n_clusters= 2, random_state=42)
```

```python
k_means.fit(df)
```

```python
clusters = k_means.labels_
clusters
```

```python
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=clusters, s=75)
```

```python
centroids = k_means.cluster_centers_
centroids
```

```python
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=clusters, s=75)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=250)
```

Elbow Methond

```python
rate = []

k = range(2, 10)

for i in k:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df)
    rate.append(kmeans.inertia_)
```

```python
plt.plot(k, rate, 'red')
plt.xlabel('K')
plt.title('Elobw')
```

```python
from yellowbrick.cluster import KElbowVisualizer
```

```python
kmeans_yellow = KMeans()
graph = KElbowVisualizer(kmeans_yellow, k=(2, 10))
graph.fit(df)
graph.poof()
```

```python
nk_means = KMeans(n_clusters= 4, random_state=42)
```

```python
nk_means.fit(df)
```

```python
new_clusters = nk_means.labels_
new_clusters
```

```python
df['cluster_id'] = new_clusters
```

```python
df.sample(10)
```

silhouette_score

```python
from sklearn.metrics import silhouette_score
```

```python
print(f"silhouette_score at (n=4) is {silhouette_score(df, new_clusters)}")
```

```python
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=new_clusters, s=75, cmap="viridis")
plt.scatter(nk_means.cluster_centers_[:, 0], nk_means.cluster_centers_[:, 1], c='red', s=250)
```
